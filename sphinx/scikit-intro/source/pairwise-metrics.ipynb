# Pairwise metrics

## Data

```python
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)
```

## Plotting function

```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt

def plot_heatmap(X, figsize=(7,7)):
    fig, ax = plt.subplots(figsize=figsize)
    sns.heatmap(X, ax=ax)
```

## Real-valued vector space

### Cosine similarity

```python
from sklearn.metrics.pairwise import cosine_similarity

S = cosine_similarity(X)
plot_heatmap(S)
```

### Cosine distance

```python
from sklearn.metrics.pairwise import cosine_distances

D = cosine_distances(X)
plot_heatmap(D)
```

### Euclidean distance

```python
from sklearn.metrics.pairwise import euclidean_distances

D = euclidean_distances(X)
plot_heatmap(D)
```

### Manhattan distance

```python
from sklearn.metrics.pairwise import manhattan_distances

D = manhattan_distances(X)
plot_heatmap(D)
```

### Kernel distance

```python
from sklearn.metrics.pairwise import pairwise_kernels

D = pairwise_kernels(X, metric='chi2')
plot_heatmap(D)
```

### Minkowski

```python
from sklearn.neighbors import DistanceMetric

D = DistanceMetric.get_metric('minkowski').pairwise(X)
plot_heatmap(D)
```

### Mahalanobis

```python
import numpy as np
from sklearn.neighbors import DistanceMetric

D = DistanceMetric.get_metric('mahalanobis', V=np.cov(X.T)).pairwise(X)
plot_heatmap(D)
```

### Haversine distance

```python
from sklearn.metrics.pairwise import haversine_distances
import pandas as pd

df = pd.read_csv('capitals.tsv', sep='\t').sample(n=10, random_state=37)
X = df[['latitude', 'longitude']].values

D = haversine_distances(X)
plot_heatmap(D, figsize=(5, 5))
```
