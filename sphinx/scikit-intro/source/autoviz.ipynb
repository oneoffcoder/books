# Autoviz

[Autoviz](https://github.com/AutoViML/AutoViz) is a visualization library.

```python
from sklearn.datasets import fetch_california_housing

X, y = fetch_california_housing(return_X_y=True, as_frame=True)
X['MedHouseVal'] = y
```

```python
from autoviz.AutoViz_Class import AutoViz_Class

AV = AutoViz_Class()

dft = AV.AutoViz(**{
    'filename': '',
    'depVar': 'MedHouseVal',
    'dfte': X,
    'chart_format': 'svg'
})
```
