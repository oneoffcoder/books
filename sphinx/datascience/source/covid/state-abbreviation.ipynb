# State Abbreviations

- [Data](https://pe.usps.com/text/pub28/28apb.htm)

```python
import pandas as pd

with open('./state-abbreviation.txt', 'r') as f:
    lines = [line.strip() for line in f.readlines() if len(line.strip()) > 0]

lines = [(lines[i], lines[i+1]) for i in range(0, len(lines), 2)]
df = pd.DataFrame(lines, columns=['name', 'abbreviation'])
df
```

```python
df.to_csv('./state-abbreviation.csv', index=False)
```
