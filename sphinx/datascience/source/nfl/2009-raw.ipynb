```python
import pandas as pd

rename_cols = {
    'Unnamed: 5': 'home_team',
    'Unnamed: 7': 'boxscore',
    'Week ': 'week',
    'Winner/tie ': 't1',
    'Loser/tie ': 't2',
    'PtsW ': 's1',
    'PtsL ': 's2',
    'TOW ': 'to1',
    'TOL': 'to2',
    'YdsW ': 'yd1',
    'YdsL ': 'yd2'
    
}
columns = ['t1', 't2', 's1', 's2', 'yd1', 'yd2', 'to1', 'to2', 'home_team', 'week']
df = pd.read_csv('./2009-raw.tsv', sep='\t').rename(columns=rename_cols)[columns]
df.t1 = df.t1.apply(lambda s: s.split(' ')[-1].strip())
df.t2 = df.t2.apply(lambda s: s.split(' ')[-1].strip())
df.home_team = df.apply(lambda r: r.t2 if pd.notna(r.home_team) and r.home_team.strip() == '@' else r.t1, axis=1)

df
```

```python
teams = sorted(list(set(df.t1) | set(df.t2)))
print(len(teams))
print(teams)
```

```python
df.to_csv('2009.csv', index=False)
```
