# Plotting

Let's learn how to plot some simple graphs.

```r
set.seed(37)
```

## Box plot

Here's an example box plot.

```r
x <- rnorm(1000, mean=0, sd=1)
y <- rnorm(1000, mean=5, sd=5)

options(repr.plot.width=4, repr.plot.height=4)

boxplot(x, y, names=c('x', 'y'), xlab='variables', ylab='value', col='gray90')
title('Box Plot')
```

We may also change the box plot to be horizontal.

```r
x <- rnorm(1000, mean=0, sd=1)
y <- rnorm(1000, mean=5, sd=5)

options(repr.plot.width=4, repr.plot.height=4)

boxplot(x, y, names=c('x', 'y'), xlab='variables', ylab='value', col='gray90', horizontal=TRUE)
title('Box Plot')
```

## Scatter plot

```r
x <- rnorm(300, mean=0, sd=1)
y <- rnorm(300, mean=5, sd=5)

options(repr.plot.width=4, repr.plot.height=4)

plot(x, y, pch=3)
abline(lm(y ~ x), lty='dotted', lwd=2, col='red')
title('Scatter Plot')
```

## Pairs plot

The `pairs plot` shows multiple scatter plots on one graphic.

```r
x <- rnorm(300, mean=0, sd=1)
y <- rnorm(300, mean=1.5, sd=2.0)
z <- 5 + 3 * x - 2 * y + rnorm(300, mean=0, sd=1)

data <- data.frame(
    x = x,
    y = y,
    z = z
)

options(repr.plot.width=5, repr.plot.height=5)

pairs(data)
```

## Line plot

```r
t <- seq(1:100)
s <- sin(t) + rnorm(100, mean=1, sd=1)

options(repr.plot.width=10, repr.plot.height=5)

plot(t, s, type='l')
```

## Pie charts

```r
e <- c(100, 50, 200)
l <- c('business', 'social', 'military')
c <- c('gray40', 'gray50', 'gray90')

options(repr.plot.width=4, repr.plot.height=4)

pie(e, labels=l, col=c)
```

## Dot charts

```r
e <- c(100, 50, 200)
l <- c('business', 'social', 'military')

options(repr.plot.width=4, repr.plot.height=4)

dotchart(e, labels=l)
```

## Bar chart

```r
e <- c(100, 50, 200)
l <- c('business', 'social', 'military')

options(repr.plot.width=4, repr.plot.height=4)

barplot(e, names=l)
```
