```r
library(dagitty)
library(lavaan)
```

```r
fig2.5 <- dagitty("dag {
    X -> R -> S -> T <- U <- V -> Y
}")
plot( graphLayout( fig2.5 ) )
```

```r
fig2.6 <- dagitty("dag {
    X -> R -> S -> T <- U <- V -> Y
    T -> P
}")
plot( graphLayout( fig2.6 ) )
```

```r
d <- simulateSEM( fig2.5, .7, .7, N=10000 )
```

```r
write.csv(d,'fig2_5.csv', row.names = FALSE)
```

```r
d <- simulateSEM( fig2.6, .7, .7, N=10000 )
```

```r
write.csv(d,'fig2_6.csv', row.names = FALSE)
```
