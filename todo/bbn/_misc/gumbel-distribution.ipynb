# Gumbel Distribution

The [Gumbel distribution](https://en.wikipedia.org/wiki/Gumbel_distribution) is used to model distributions of maximum or minimum values.

$E(X) = \mu + \gamma \beta$, where $\gamma$ is the Euler-Mascheroni constant ($\gamma = 0.57721$).

```python
import numpy as np
import pandas as pd
from numpy.random import gumbel

np.random.seed(37)

N = 10_000
means = [0.0, 0.5, 1, 1.5, 3.0]
df = pd.DataFrame({f'm={m}': gumbel(loc=m, scale=1, size=10_00)for m in means})
```

```python
df.plot(kind='kde')
```

```python
df.describe()
```

```python
df.describe().loc['mean'] - 1 * 0.57721
```
